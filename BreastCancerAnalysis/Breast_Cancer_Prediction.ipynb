```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/BreastCancer.csv')
```

```python
data.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':0,'B':1})
```

```python
data.head()
```

```python
data.shape
```

1 ---> Benign

0 ---> Malignant

```python
data['diagnosis'].value_counts()
```

```python
data.groupby('diagnosis').mean()
```

```python
# seperating the features and target
X = data.drop(columns='diagnosis',axis=1)
Y = data['diagnosis']
```

```python
X
```

```python
Y
```

```python
#Splitting the data into trainin data and testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape , X_train.shape, X_test.shape)
```

```python
#Model training
#Logistic Regression

model = LogisticRegression()
```

```python
# training the logistic regression ,odel using training data
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
#Accuracy on training data

X_pred = model.predict(X_train)

Accuracy_score1 = accuracy_score(Y_train,X_pred)
```

```python
print(f'Accuracy score of the data : {Accuracy_score1}')
```

```python
#Accuracy on testing data

Y_pred =  model.predict(X_test)

Accuracy_score2 = accuracy_score(Y_test,Y_pred)
```

```python
print(f'Accuracy score of the data : {Accuracy_score2}')
```
